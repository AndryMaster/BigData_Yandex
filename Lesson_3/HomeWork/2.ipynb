```python
import re
import pandas as pd

pd.options.mode.copy_on_write = True


def trans_time(time: str) -> int:
    # time = time.lstrip('c').lstrip(':')
    # d = ':'
    # if "'" in time:
    #     d = "'"
    # elif '.' in time:
    #     d = '.'
    # 
    # time = time.split(d)  # ':'
    # time = int(time[0]) * 60 + int(time[1])
    
    # try:
    #     time = re.sub('[^0-9]',' ', time).split()
    #     time = int(time[0]) * 60 + int(time[1])
    # except Exception as e:
    #     return 0
    
    time = re.sub('[^0-9]', ' ', time).split()
    time = int(time[0]) * 60 + int(time[1])

    if 8 * 60 <= time <= 10 * 60:
        return -1
    elif 20 * 60 <= time <= 22 * 60:
        return 1
    return 0


def process(df: pd.DataFrame):
    df = df[['Time', 'Fatalities']].dropna(subset=['Time'], how='all')
    df['TimeCat'] = df['Time'].apply(lambda s: trans_time(s))
    # print(df.loc[df['TimeCat'] == -1])
    # print(df.loc[df['TimeCat'] == 1])
    return round(df.loc[df['TimeCat'] == -1]['Fatalities'].sum() / df.loc[df['TimeCat'] == 1]['Fatalities'].sum(), 2)
```

```python
df = pd.read_csv("airplane_crushes.csv")

df.info()
df.describe()

process(df)
```
