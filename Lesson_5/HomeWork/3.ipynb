## Исправляем прогноз

```python
import matplotlib.pyplot as plt
from sklearn import metrics
import pandas as pd
import numpy as np


def process(df: pd.DataFrame):
    df.prediction = 1.0 - df.prediction
    return df
```

```python
df = pd.read_csv('classifier_prediction_2.csv')
df.prediction = 1.0 - df.prediction

fpr, tpr, thresholds = metrics.roc_curve(y_true=df.target, y_score=df.prediction)
roc_auc = metrics.roc_auc_score(y_true=df.target, y_score=df.prediction)

print(roc_auc)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.grid()
plt.show()
```
