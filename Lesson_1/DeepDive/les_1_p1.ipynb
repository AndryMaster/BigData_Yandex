## JSON в CSV на минималках

```python
import json
import csv

data = json.load(open('cases.json'))
data = sorted(data, key=lambda k: k['date'], reverse=True)
data = [dict((k, v) for k, v in d.items() if k != 'description') for d in data]
print(data)

with open("incidents.csv", "w", newline="") as f:
    w = csv.DictWriter(f, "creature,place,danger,date".split(','))
    w.writeheader()
    w.writerows(data)
```

## Уникализация на минималках

```python
import re

with open("ticket_logs.csv", "r", encoding='utf-8') as f:
    data = f.readlines()
    
# res = [re.sub('\D', '', line.split(',')[1]) for line in data]
# print(res)
res = {}
count = {}
for row in data:
    title, phone = row.split(',')
    phone = re.sub('\D', '', phone)
    if len(phone) != 11:
        continue
    
    if title in res:
        res[title].append(phone)
    else:
        res[title] = [phone]
    
    count[title] = count.get(title, 0) + 1

max_title = max(count, key=count.get)
# print(count)
# print(max(count.values()))
# print(len(set(re.sub('\D', '', line.split(',')[1]) for line in data)))
print(len(set(res[max_title])))
```

## Склейка на минималках

```python
with open("transaction_logs.csv", "r", encoding='utf-8') as f:
    data = [row.strip().split(',') for row in f.readlines()]

cidx = {}
idx = 1
n_components = 0
prev_n_components = -1

while n_components != prev_n_components:
    prev_n_components = n_components
    for row in data:
        u, v = row[0], row[1]
        if u in cidx and v in cidx:
            if cidx[u] != cidx[v]:
                newidx = min(cidx[u], cidx[v])
                cidx[u] = newidx
                cidx[v] = newidx
                n_components -= 1
        elif u in cidx:
            cidx[v] = cidx[u]
        elif v in cidx:
            cidx[u] = cidx[v]
        else:
            cidx[u] = idx
            cidx[v] = idx
            n_components += 1
            idx += 1

output = [(u, v, cidx[u]) for u, v in data]
# print(*output, sep='\n')
# print(cidx, sep='\n')

res = {}
for u, v in data:
    res[cidx[v]] = res.get(cidx[v], 0) + 1

print(max(res.values()))
```
