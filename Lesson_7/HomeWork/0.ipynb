```python
from IPython.display import display
import pandas as pd
import numpy as np


df = pd.read_csv('plan_fact.csv')

display(df.describe())
display(df.head())
df.info()
print(df.nunique())
print(df.count())
```

```python
for m in df.manager.unique():
    print(m, df.loc[df.manager == m].segment.unique())
```

```python
df['value_fact_nak'] = df['value_fact']
df['value_plan_nak'] = df['value_plan']

for name in ['value_fact_nak', 'value_plan_nak']:
    for i in range(1, df.shape[0]):
        df.loc[i, name] += df.loc[i-1, name]
    
df.to_csv('plan_fact_nak.csv')
```

```python
df['value_fact_nak_2022'] = df['value_fact']

for i in range(df.shape[0]):
    if df.loc[i, 'date'].startswith('2022-'):
        df.loc[i, 'value_fact_nak_2022'] += df.loc[i-1, 'value_fact_nak_2022']
    else:
        df.loc[i, 'value_fact_nak_2022'] = 0
```

```python
# df.head(10)
df.to_csv('plan_fact_nak.csv', index=False)
```

```python
df = df.assign(plan_fact=pd.Series(np.zeros(shape=df.shape[0])).values)
df['plan_fact_nak'] = df['plan_fact']
display(df.head(7))
# df['plan_fact'] = df['value_fact'] / df['value_plan']
# df['plan_fact_nak'] = df['plan_fact']

for i in range(1, df.shape[0]):
    if df.loc[i, 'date'].startswith('2022-') and df.loc[i, 'value_plan'] > 0:
        df.loc[i, 'plan_fact'] = df.loc[i, 'value_fact'] / df.loc[i, 'value_plan']
        df.loc[i, 'plan_fact_nak'] = df.loc[i-1, 'plan_fact_nak'] + df.loc[i, 'plan_fact']

display(df.head(10))
display(df.tail(10))
```
