## Нечестная монетка - 2

```python
import numpy as np


def process(n: int) -> float:
    coins = np.random.choice([0, 1], n, p=[0.4, 0.6])
    res = []
    for i in range(1000):
        sample = np.random.choice(coins, size=n, replace=True)
        res.append(sample.sum() / sample.size)
    res.sort()
    # print(f'{n=}: {np.mean(res)}')
    return res[-26] - res[25]
```

```python
import matplotlib.pyplot as plt

pres = []
for n in [1000, 5000, 10000, 50000]:
    pres.append([n, process(n)])

print(pres)
plt.figure(figsize=(10, 6))
plt.plot(*zip(*pres))
plt.grid()
plt.show()
```
