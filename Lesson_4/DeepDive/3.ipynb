```python
import pandas as pd
pd.options.mode.copy_on_write = True


def process(df: pd.DataFrame):
    df_find = df[['Name', 'Age', 'Weight', 'Height', 'Overall']]

    df_find['Height'] = df_find['Height'].apply(lambda s: int(s.split("'")[0]) * 0.3048 + int(s.split("'")[1]) * 0.0254)
    df_find['Weight'] = df_find['Weight'].apply(lambda s: float(s[:-3]) * 0.453592)
    df_find['BMI'] = df_find['Weight'] / df_find['Height'] ** 2
    df_mean = df_find.mean(numeric_only=True)
    df_find['Normality'] = sum([(df_find[col_name] - df_mean[col_name]) ** 2
                                for col_name in ['Age', 'Weight', 'Height', 'Overall', 'BMI']]) ** 0.5
    df_find.sort_values(by='Normality', ascending=True, inplace=True)
    
    return df_find.head(3).reset_index(drop=True)[['Weight', 'Height', 'Age', 'Overall', 'BMI', 'Name', 'Normality']]
```

```python
df = pd.read_csv("FIFA21_official_data.csv")

df.info()
df.describe()
```

```python
process(df)
```
