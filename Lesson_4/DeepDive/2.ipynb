```python
import pandas as pd


def sort_reset(df: pd.DataFrame):
    df.sort_values(by=['ID'], ascending=True).reset_index(drop=True, inplace=True)


def process(df1: pd.DataFrame, df2: pd.DataFrame):
    sort_reset(df1)
    sort_reset(df2)
    df = df1.combine_first(df2).dropna(axis=0, how='any').sort_values(by=['ID'], ascending=True).reset_index(drop=True)
    return df[[col for col in df if col != 'ID'] + ['ID']]


def process_(df1: pd.DataFrame, df2: pd.DataFrame):
    return df1.combine_first(df2).dropna(axis=0, how='any', inplace=True).sort_values(by=['ID'], ascending=True, inplace=True)
```

```python
df1 = pd.read_csv("FIFA22_broken.csv")
df2 = pd.read_csv("FIFA23_broken.csv")

df1.info()
df1.describe()
```

```python
df3 = process(df1, df2)

print(df3.iloc[:10, 0])
df3.info()
df3.describe()
```

```python
df1.sort_values(by=['ID'], ascending=True, inplace=True)
df2.sort_values(by=['ID'], ascending=True, inplace=True)
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

print(df1['ID'].head(10))
print(df2['ID'].head(10))
print(df1['ID'].count())
print(df2['ID'].count())
```
