## RFM-скоринг

```python
import pandas as pd

pd.options.mode.copy_on_write = True


def process(df: pd.DataFrame):
    df['R_rank'] = df['Recency'].rank(ascending=False)
    df['R_norm_rank'] = df['R_rank'] / df['R_rank'].max()
    
    df['F_rank'] = df['Frequency'].rank(ascending=True)
    df['F_norm_rank'] = df['F_rank'] / df['F_rank'].max()
    
    df['M_rank'] = df['Monetary'].rank(ascending=True)
    df['M_norm_rank'] = df['M_rank'] / df['M_rank'].max()
    
    df = df[['Customer_ID', 'R_norm_rank', 'F_norm_rank', 'M_norm_rank']]
    df['RFM_score'] = 5 * (15 * df['R_norm_rank'] +
                           28 * df['F_norm_rank'] +
                           57 * df['M_norm_rank']) / 100
    
    df['select'] = df['RFM_score'].apply(lambda n: 4 <= n <= 4.5)
    res = df.loc[df['select']]['Customer_ID'].to_list()

    return res
```
