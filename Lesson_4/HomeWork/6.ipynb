## Confidence для Apriori

```python
import itertools
import csv

min_support = float(input())
min_confidence = float(input())

all_products = set()
operations = []
count_one = dict()

with open('store_data.csv') as file:
    for line in file.readlines():
        line = set(line.strip().split(','))
        operations.append(line)
        for el in line:
            all_products.add(el)
            count_one[el] = count_one.get(el, 0) + 1
            
print(all_products)
print(len(operations))

total_count = len(operations)
support = {}
confidence = {}
res = []

for comb in itertools.combinations(all_products, 2):
    # print(comb)
    count = 0
    for line in operations:
        if comb[0] in line and comb[1] in line:
            count += 1
    
    support_comb = count / total_count
    confidence1 = count / count_one[comb[0]]
    confidence2 = count / count_one[comb[1]]
    if support_comb >= min_support:
        if confidence1 >= min_confidence:
            res.append([comb[0], comb[1], round(confidence1, 3)])
        if confidence2 >= min_confidence:
            res.append([comb[1], comb[0], round(confidence2, 3)])
        
    support[comb] = support_comb
    confidence[comb] = confidence1
    confidence[tuple(reversed(comb))] = confidence2
    
print(support)
print(confidence)

with open('output.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow('first_item second_item confidence'.split())
    writer.writerows(res)
```
