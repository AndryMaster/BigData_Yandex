## Support для Apriori

```python
import itertools
import csv

min_support = float(input())

all_products = set()
operations = []
support = {}
res = []

with open('store_data.csv') as file:
    for line in file.readlines():
        line = set(line.strip().split(','))
        operations.append(line)
        for el in line:
            all_products.add(el)
            
# print(all_products)
# print(len(operations))
total_count = len(operations)

for comb in itertools.combinations(all_products, 2):
    count = 0
    for line in operations:
        if comb[0] in line and comb[1] in line:
            count += 1
    
    support_comb = count / total_count
    if support_comb >= min_support:
        res.append([comb[0], comb[1], round(support_comb, 3)])
        res.append([comb[1], comb[0], round(support_comb, 3)])
    # support[comb] = support_comb
    
# print(support)

with open('output.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow('first_item second_item support'.split())
    writer.writerows(res)
```
